#Shuffling randomly for providing robustness

```python
import random
random.shuffle(password_tuple)
```

```python
x=[labels[0] for labels in password_tuple]
y=[labels[1] for labels in password_tuple]
```

# Creating a custom function to split input into characters of list

```python
def word_divide_char(inputs):
    character=[]
    for i in inputs:
        character.append(i)
    return character
```

```python
word_divide_char('kzde5577')
```

#Importing TF-IDF vectorizer to convert String data into numerical data

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer=TfidfVectorizer(tokenizer=word_divide_char)
```

# Apply TF-IDF vectorizer on data

```python
X=vectorizer.fit_transform(x)
```

```python
X.shape
```

```python
vectorizer.get_feature_names()
```

```python
first_document_vector=X[0]
first_document_vector
```

```python
first_document_vector.T.todense()
```

```python
df=pd.DataFrame(first_document_vector.T.todense(),index=vectorizer.get_feature_names(),columns=['TF-IDF'])
df.sort_values(by=['TF-IDF'],ascending=False)
#it is sorted according to its importance
```

#Spliting data into train & test
    

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

#Applying Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression(random_state=0,multi_class='multinomial') # becoz dependent variable has more than 2 categories 
```

```python
clf.fit(X_train,y_train)
```

Prediction on X-Test data

```python
y_pred=clf.predict(X_test)
y_pred
```

#Predicting for random custom data

```python
dt=np.array(['%@123abcd'])
pred=vectorizer.transform(dt)
clf.predict(pred)
```

So the above password is in average strength

#Checking Accuracy of the model using confusion_matrix and accuracy_score

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(accuracy_score(y_test,y_pred))
```

#Classification_report




```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
